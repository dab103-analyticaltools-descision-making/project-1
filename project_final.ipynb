```python
ptds=ds[ds["commodity"]=="Potato"]
ptds_sum=ptds.groupby(['state']).sum().sort_values("modal_price")
ptds_sum
```

```python
xp=ptds_sum.index
yp=ptds_sum.modal_price
plt.figure(figsize=(8,8),dpi=80)
plt.plot(xp,yp)
plt.xticks(rotation=90)
plt.title("revenue of each state getting from potato in year 2021")
plt.xlabel("states")
plt.ylabel("revenue from potato(crores)")
plt.show()
```

**Above plot shows the revenue generated from potato in each state where as  meghalaya marked lowest and uttar pradesh marked as highes revenue from potato**

```python
rcds =ds[ds["commodity"]=="Rice"]
rcds_sum=rcds.groupby(['state']).sum().sort_values("modal_price")
rcds_sum
```

```python
xp=rcds_sum.index
yp=rcds_sum.modal_price
plt.figure(figsize=(8,8),dpi=80)
plt.plot(xp,yp)
plt.xticks(rotation=90)
plt.title("revenue of each state getting from Rice in year 2021")
plt.xlabel("states")
plt.ylabel("revenue from rice(crores)")
plt.show()
```
