```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
ds=pd.read_csv("B:\\analytic tools and decision making 103\project\merged file.csv")
ds
```

```python
ds[ds['min_price']==0].count()[1]
```

**Eventough there are 28 zero values in min_price it's fine. because minimum can start from zero.**

```python
ds[ds['max_price']==0].count()[1]
```

**there are 187 zero values which are invalid because min_price can be zero but Max_price cannot be zero. But we need to see that is there any zero values for both min_price and max_price**

```python
ds=ds[(ds['max_price']!=0) & (ds['min_price']!=0)]
ds
```

as there are zero values in both minimum and maximum prices which has no meaning so they are considered as missing values

```python
ds=ds[ds['variety']!="Other"]
ds
```

There are many values named other which are also considered as missing values in our analysis

```python
pd.options.mode.use_inf_as_na = True
ds.isna()
```

**this table has no NA values**

```python
ds=ds[ds['variety']!="F.A.Q."]
ds
```

In variety there is a value **F.A.Q.** which is considered as invalid value

##Rice##

```python
rcds =ds[ds["commodity"]=="Rice"]
rcds_sum=rcds.groupby(['state']).sum().sort_values("modal_price")
rcds_sum
```

```python
xp=rcds_sum.index
yp=rcds_sum.modal_price
plt.figure(figsize=(8,8),dpi=80)
plt.plot(xp,yp)
plt.xticks(rotation=90)
plt.title("revenue of each state getting from Rice in year 2021")
plt.xlabel("states")
plt.ylabel("revenue from rice(crores)")
plt.show()
```

**Above plot shows the revenue generated from rice in each state where as  Manipur marked lowest and uttar pradesh marked as highes revenue from rice**

```python

rcvar=rcds.groupby(['variety']).sum().sort_values("modal_price")
rcvar
```

```python
x=rcvar.index
y=rcvar.modal_price
plt.figure(figsize=(12,8),dpi=80)
plt.plot(x,y)
plt.xticks(rotation=90)
plt.title("which variety of rice is generating huge revenue")
plt.xlabel("variety of Rice")
plt.ylabel("revenue from potato(millions)")
plt.show()
```

we can see **common, Fine, III** are three varieties that are generating huge amount of revenue

```python
rcst=rcds[(rcds['variety']=="Common") | (rcds['variety']=="Fine")| (rcds['variety']=="III")]
rcst
```

```python

sns.catplot(x='state', hue="variety", kind="count", height=8.27, aspect=11.7/8.27,data=rcst).set(title="this shows the ditribution of states cultivating high revenue generated rice.")
```

We saw in previous plots that uttar pradesh had generated huge revenue. now Comparatively, Uttar Pradesh Is cultivating **III** variety of rice and transporting to every part of india.

```python
rccost=rcds.groupby(["variety"]).mean().sort_values("modal_price")
rccost.astype(int)
```

```python
rccost['variety']=rccost.index
sns.catplot(x='variety',y="modal_price",kind="bar",height=8.27, aspect=16.7/9.27,data=rccost).set(title="this shows that the average cost of each variety")
plt.xticks(rotation=90)
plt.show()
```

Above plot shows that the average cost of each varity of rice. moreover, this shows that **Basumathi, basmati and Fine** are three variety of rice which are costly.because, they are raised in few states. furthermore, the demand is high and supply is low. so the prices of these 3 variety are at peaks. however, In previous plots we saw that uttar pradesh is generating huge revenue by farming **III** variety. in this senario, the **III** variety has high supply and high demand. as a result, tough the price of **III** is low utaar pradesh has high revenue.
